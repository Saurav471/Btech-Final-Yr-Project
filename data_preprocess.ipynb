```python
f = open('data.csv','r')
data = f.read()
f.close()

data = data.split('\n')
```

```python
final_data = []
for index, row in enumerate(data):
    row = row.split('\t')
    final_data.append(row)
```

```python
import pandas as pd

df = pd.DataFrame(final_data)
df.head()
```

```python
df.to_csv('final_data.csv', index=False, header=False)
```

```python
df = pd.read_csv('final_data.csv')
df.head()
```

```python
for i in df.columns:
    if 'A' not in i:
        df = df.drop([i], axis = 1)

df.head()
```

```python
#anxiety 2,4,41,28,25,40,20
#depression 3,42,10,26,31,38
#stress 6,22,12,39,8,35,18
a = ['Q2A','Q4A','Q41A','Q28A','Q25A','Q40A','Q20A']
d = ['Q3A','Q42A','Q10A','Q26A','Q31A','Q38A']
s = ['Q6A','Q22A','Q12A','Q39A','Q8A','Q35A','Q18A']

a_df = df[a]
a_df['sum_a'] = a_df.iloc[:,0:].sum(axis=1)
df['sum_a'] = a_df['sum_a']

d_df = df[d]
d_df['sum_d'] = d_df.iloc[:,0:].sum(axis=1)
df['sum_d'] = d_df['sum_d']

s_df = df[s]
s_df['sum_s'] = s_df.iloc[:,0:].sum(axis=1)
df['sum_s'] = s_df['sum_s']

df.head()
```

```python
def label_fn (row):
    if (row['sum_a'] >= row['sum_d']) and (row['sum_a'] >= row['sum_s']):
        return 'anxiety'
    elif (row['sum_d'] >= row['sum_a']) and (row['sum_d'] >= row['sum_s']):
        return 'depression'
    else:
        return 'stress'
```

```python
df['label'] = df.apply(lambda row: label_fn(row), axis=1)
df.head()
```

```python
df = df.drop(['sum_a', 'sum_d', 'sum_s'], axis = 1)
df.to_csv('labeled_data.csv', index=False)
```
